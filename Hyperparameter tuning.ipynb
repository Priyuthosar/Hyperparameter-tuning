```python
import pandas as pd
import numpy as np
```

```python
from sklearn import svm, datasets
iris = datasets.load_iris()
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
df['target']=iris.target
```

```python
df['target'] = df['target'].apply(lambda x:iris.target_names[x])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(iris.data,iris.target,test_size=0.25)
```

```python
model= svm.SVC(kernel='rbf',C=30,gamma='auto')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(svm.SVC(kernel='linear',C=10,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
cross_val_score(svm.SVC(kernel='rbf',C=20,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
cross_val_score(svm.SVC(kernel='linear',C=20,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
svm.SVC()
```

```python
kernels = ['rbf','linear']
C=[1,10,20,30]
avg_scores = {}
for kval in kernels:
    for cval in C:
        cv_scores = cross_val_score(svm.SVC(kernel = kval,C=cval, gamma='auto'),iris.data,iris.target,cv=5)
        avg_scores[kval +'_'+ str(cval)] = np.average(cv_scores)
avg_scores
```

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'),{
    'C': [1,5,10,20,30],
    'kernel': ['rbf','linear','sigmoid','poly'],
    'max_iter':[1,5,10,20,100]   },cv =10, return_train_score = False)
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_C','param_kernel','param_max_iter','mean_test_score']]
```

```python
clf.best_params_
```

```python
clf.best_score_
```

using different models with different hyperparameters , will find which is the best model to consider

```python
from sklearn import  svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
```

```python
model_params = {
    
    'svm':{
        'model':svm.SVC(gamma='auto'),
        'params':{
            'C':[1,10,20,30,50],
            'kernel':['rbf','linear']
        }
    },
    'random_forest':{
        'model': RandomForestClassifier(),
        'params':{
            'n_estimators':[1,5,10,20,50]
        }
    },
    'logistic_regression':{
        'model':LogisticRegression(solver='liblinear',multi_class='auto'),
        'params':{
            'C': [1,5,10,20]
        }
    }  
    
}
```

```python
scores =[]
for model_name,mp in model_params.items():
    clf = GridSearchCV(mp['model'],mp['params'],cv=10,return_train_score=False)
    clf.fit(iris.data,iris.target)
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    
    })
dfinal = pd.DataFrame(scores,columns=['model','best_score','best_params'])
dfinal
```
